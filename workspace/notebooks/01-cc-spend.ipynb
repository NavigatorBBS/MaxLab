# 3 â€” Data Ingestion, Cleaning, & Normalization

### Credit Card transformation

```python
my_cc_acct_df = pd.read_csv(RAW_DATA_DIR + my_cc_acct_csv)
print(my_cc_acct_df.columns)
my_cc_acct_df = my_cc_acct_df.drop(columns=['Account Number','Cheque Number','Description 2','USD$'])
my_cc_acct_df = my_cc_acct_df.rename(columns={
    "Account Type": "Type",
    "Transaction Date": "Date",
    "Description 1": "Description"
})
my_cc_acct_df['Date'] = pd.to_datetime(my_cc_acct_df['Date'])
my_cc_acct_df['Date'] = my_cc_acct_df['Date'].dt.strftime('%d-%m-%Y')
my_cc_acct_df['Debit'] = my_cc_acct_df['CAD$'].where(my_cc_acct_df['CAD$'] < 0, 0).abs()
my_cc_acct_df['Credit'] = my_cc_acct_df['CAD$'].where(my_cc_acct_df['CAD$'] > 0, 0)
my_cc_acct_df = my_cc_acct_df.drop(columns=['CAD$'])
my_cc_acct_df.head()
```
